```python
import matplotlib.pyplot as plt
import numpy as np



```

```python
def f(t):
    
    """
    t*np.cos(t) + 1
    """
    
    return t*np.cos(t) + 1
```

```python
def bisection(tolerance, xl, xu):
    iteration = 0
    e = (xu - xl) / 2
    
    while e > tolerance:
        t1 = np.arange(xl, xu, 0.1)
        c = (xl + xu) / 2
        fxlc = f(xl) * f(c)
        fxuc = f(xu) * f(c)
        
        if fxlc > 0:
            xl = c
        elif fxuc > 0:
            xu = c
            
        e = (xu - xl) / 2
        #plt.plot(t1, f(t1), "-gD", markevery=c)
        print(f"[+] Iteration `{iteration}` Error `{e}` Midpoint `{c}`")
        iteration = iteration + 1
```

```python
def false_position(tolerance, xl, xu):
    iteration = 0
    if xl * xu > 0.0:
        print("[-] Given guess values do not bracket the root")
        return
    e = (xu - xl) / 2
    
    while e > tolerance:
        c = xu - ((xl*f(xu)) - (xu*f(xl))) / (f(xu) - f(xl))
        fxlc = f(xl) * f(c)
        fxuc = f(xu) * f(c)
        
        if f(c) == 0:
            print("[+] The root is: ", c)
        elif fxlc < 0:
            xu = c
        else:
            xl = c

        e = abs(f(c))
        #plt.plot(t1, f(t1), "-gD", markevery=c)
        print(f"[+] Iteration `{iteration}` Error `{e}` Midpoint `{c}`", xl, xu, fxlc, fxuc)
        iteration = iteration + 1
```
